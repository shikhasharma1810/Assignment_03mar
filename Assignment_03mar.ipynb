```python
1. Pair Plot: Plot pairwise relationships in a dataset.
2. Joint Plot: Draw a plot of two variables with bivariate and univariate graphs.
3. Display Plot: Figure-level interface for drawing distribution plots onto a FacetGrid.
4. Relational Plot: Figure-level interface for drawing relational plots onto a FacetGrid.
5. Categorical Plot: Figure-level interface for drawing categorical plots onto a FacetGrid.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
# Kind of plot to draw, corresponding to a seaborn relational plot.
# Options are `"scatter"` or `"line"`.
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line')
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
import matplotlib.pyplot as plt

# Pivot the dataset to have months as rows and years as columns
flights_pivoted = flights.pivot("month", "year", "passengers")
# Plot the heatmap using the pivoted dataset
# Plot heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(flights_pivoted, annot=True, cmap='coolwarm', fmt=".2f",annot_kws={"size": 8})
# Show the plot 
plt.show()
```
